# Linear Algebra 

Learning linear algebra helps us solve real world problems. We begin with a simple 2 variable problem problem and slowing work our way up to much more complex solutions.

> What is linear algebra.

## Solving a 2 Variable problem will require 2 equations
1. y = 2x + 5 # is a linear algebraic equation

Simplified => y - 2x = 5

With 2 linear equations we can solve for x and y.

Lets add another equation:

2. y - 3x = 4
Simplified => y - 3x = 4

These 2 equations can be represented as a Matrix. See below.


## Vector manipulation using Gausian Elimination.

> Dot Operations

Algebraically, the dot product is the sum of the products of the corresponding entries of the two sequences of numbers.


```python
import numpy as np
A = np.array([[1, -2],[1, -3]])
C = np.array([5, 4])
B = np.linalg.solve(A, C)
B
```

> Inverse Operations

A is Matrix
A' is the inverse of the Matrix A

> Inverse Rule ==> [A]' dot [A] = 1

```python
AInverse = np.linalg.inv(A)
AInverse
```

> Inverse Operation to Find B

1. Let, [A] dot [B] = [C]

> Multiply both sides with [A]'

2. [A]' dot [A] dot [B] = [A]' dot [C]

> Use the Inverse Rule

3.   1 dot [B] = [A]' dot [C]
4.   [B] = [A]' dot [C]




```python
BCheck = np.multiply(AInverse,C)
BCheck
```
